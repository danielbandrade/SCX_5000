```python
# Lecture 3 - HW5

def first_avarage( S_1 , S_2 ):

  if( (S_1 < 0 or S_1 > 10) or (S_2 < 0 or S_2 > 10) ):
    return "Invalid Score"
  else:
    return ( S_1 + S_2 ) / 2

S_1 = float( input( "Enter the score of the first exam: " ))
S_2 = float ( input( "Enter the score of the second exam: " ))

first_avarage = first_avarage( S_1 , S_2 )

if ( first_avarage == "Invalid Score" ):
  print("Invalid Score")
elif( first_avarage >= 7 ):
  print("You have passed with grade" , first_avarage)
else:
  S_3 = float ( input( "Enter the score for the third exam: "))
  if(S_3 < 0 or S_3 > 10 ):
    print("Invalid Score")
  else:
    if ( S_3 >= 5):
      print("You have passed the course")
    else:
      print("You have failed")

```
