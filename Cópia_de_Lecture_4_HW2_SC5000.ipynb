```python
def numbers_between_2a_2b(a,b):
  two_a = 2 * a
  two_b = 2 * b
  for i in range(two_a,two_b+1):
    print(i)

a = int(input("Type a: "))
b = int(input("Type b: "))

numbers_between_2a_2b(a,b)
```
